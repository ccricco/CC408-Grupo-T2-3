```python
# === 1. Cargar base ===
df_matriz = df_filtrado.copy()   # <-- asegurate de que el archivo esté en tu carpeta de trabajo

# === 2. Seleccionar variables de interés ===
cols_req = ["ANO4","CH04","CH06","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC","IPCF"]
df_matriz = df_matriz[cols_req].copy()

# === 3. Armonizar variables ===
# Sexo (CH04) → Varón/Mujer
def norm_sexo(x):
    if pd.isna(x):
        return np.nan
    if str(x).strip() in ["1","Varón","varon","varón"]:
        return "Varón"
    if str(x).strip() in ["2","Mujer","mujer"]:
        return "Mujer"
    return str(x)

df_matriz["CH04"] = df_matriz["CH04"].apply(norm_sexo)

# Edad (CH06) → numérica
def to_float_age(v):
    try:
        return float(v)
    except:
        s = str(v).lower()
        if "menos de 1" in s:
            return 0.5
        return np.nan

df_matriz["CH06"] = df_matriz["CH06"].apply(to_float_age)

# Ingreso (IPCF) → numérico
df_matriz["IPCF"] = pd.to_numeric(df_matriz["IPCF"], errors="coerce")

# === 4. Crear dummies para categóricas ===
cat_vars = ["CH04","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC"]
num_vars = ["CH06","IPCF"]

dummies = pd.get_dummies(df_matriz[cat_vars], drop_first=True, dtype=int)

# === 5. Dataset final para correlación ===
X = pd.concat([df_matriz[num_vars], dummies], axis=1)

# Renombrar columnas (opcional, para legibilidad)
rename_map = {}
for c in X.columns:
    new = c
    new = new.replace("CH04_", "Sexo: ")
    new = new.replace("CH07_", "Estado civil: ")
    new = new.replace("CH08_", "Cobertura: ")
    new = new.replace("NIVEL_ED_", "Educación: ")
    new = new.replace("ESTADO_", "Actividad: ")
    new = new.replace("CAT_INAC_", "Inactividad: ")
    if new == "CH06": new = "Edad (años)"
    if new == "IPCF": new = "Ingreso per cápita fam."
    rename_map[c] = new

X = X.rename(columns=rename_map)

# === 6. Calcular matriz de correlación ===
corr = X.corr(method="pearson")

# Mostrar resultados
corr
```

```python
# ==========================================
# Correlación 2005 vs 2025 con diccionario 2025
# ==========================================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re

# --- Parámetros ---
RUTA = "df_filtrado.xlsx"   # ajustá si tu archivo está en otra carpeta
ANIOS = [2005, 2025]
TOP_K = 30                  # para que el heatmap sea legible (top variables por |corr|)

# --- 1) Cargar y seleccionar ---
df = pd.read_excel(RUTA)
cols_req = ["ANO4","CH04","CH06","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC","IPCF"]
df = df[[c for c in cols_req if c in df.columns]].copy()

# --- 2) Normalizar básicos ---
# Sexo (unificar 1/2 y texto)
def norm_sexo(x):
    if pd.isna(x): return np.nan
    s = str(x).strip().lower()
    if s in ["1","varon","varón"]: return "Varón"
    if s in ["2","mujer"]:         return "Mujer"
    return str(x).strip()

# Edad (numérica, "menos de 1 año" -> 0.5)
def to_float_age(v):
    try:
        return float(v)
    except:
        s = str(v).lower()
        return 0.5 if "menos de 1" in s else np.nan

df["CH04"] = df["CH04"].apply(norm_sexo)
df["CH06"] = df["CH06"].apply(to_float_age)
df["IPCF"] = pd.to_numeric(df["IPCF"], errors="coerce")

# --- 3) DICCIONARIO OFICIAL 1T2025 (INDEC) ---
map_CH07 = {  # Estado civil
    1: "Unido", 2: "Casado", 3: "Separado o divorciado",
    4: "Viudo", 5: "Soltero"
}
map_CH08 = {  # Cobertura médica (incluye combinaciones)
    1: "Obra social/PAMI",
    2: "Mutual/Prepaga/Servicio de emergencia",
    3: "Planes y seguros públicos",
    4: "No paga ni le descuentan",
    9: "Ns/Nr",
    12: "Obra social + Mutual/Prepaga/Servicio de emergencia",
    13: "Obra social + Planes/seguros públicos",
    23: "Mutual/Prepaga/Serv. emergencia + Planes/seguros públicos",
    123: "Obra social + Mutual/Prepaga/Serv. emergencia + Planes/seguros públicos",
}
map_NIVEL_ED = {  # Nivel educativo
    1: "Primaria Incomp. (incluye educación especial)",
    2: "Primaria Comp.",
    3: "Secundaria Incomp.",
    4: "Secundaria Comp.",
    5: "Superior/Univ. Incomp.",
    6: "Superior/Univ. Comp.",
    7: "Sin instrucción",
    9: "Ns/Nr",
}
map_ESTADO = {  # Condición de actividad
    0: "Entrevista no realizada",
    1: "Ocupado",
    2: "Desocupado",
    3: "Inactivo",
    4: "Menor de 10 años",
}
map_CAT_INAC = {  # Categoría de inactividad
    1: "Jubilado/Pensionado",
    2: "Rentista",
    3: "Estudiante",
    4: "Ama de casa",
    5: "Menor de 6 años",
    6: "Discapacitado",
    7: "Otros",
}

# --- 4) Aplicar mapeos SOLO donde haya códigos (2025) sin romper 2005 en texto ---
def safe_replace(series, mapa):
    """Reemplaza solo si el valor es convertible a número y está en el diccionario."""
    def _map(v):
        try:
            k = int(float(v))
            return mapa.get(k, v)
        except:
            return v
    return series.apply(_map)

df["CH07"] = safe_replace(df["CH07"], map_CH07)
df["CH08"] = safe_replace(df["CH08"], map_CH08)
df["NIVEL_ED"] = safe_replace(df["NIVEL_ED"], map_NIVEL_ED)
df["ESTADO"] = safe_replace(df["ESTADO"], map_ESTADO)
df["CAT_INAC"] = safe_replace(df["CAT_INAC"], map_CAT_INAC)

# --- 5) Helpers de etiquetas y plotting ---
cat_vars = ["CH04","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC"]
num_vars = ["CH06","IPCF"]

def limpiar_texto(s):
    s = re.sub(r"\s+", " ", s)
    s = s.replace("Obra social (incluye PAMI)", "Obra social/PAMI")
    s = s.replace("Mutual/Prepaga/Servicio de emergencia", "Prepaga/serv. emergencia")
    s = s.replace("Planes y seguros públicos", "Planes/seguros públicos")
    s = s.replace("Universitaria", "Univ.")
    s = s.replace("Primaria", "Prim.")
    s = s.replace("Secundaria", "Sec.")
    s = s.replace("Completa", "Comp.")
    s = s.replace("Incompleta", "Incomp.")
    return s

def renombrar_columnas(X):
    rename_map = {}
    for c in X.columns:
        new = c
        new = new.replace("CH04_", "Sexo: ")
        new = new.replace("CH07_", "Estado civil: ")
        new = new.replace("CH08_", "Cobertura: ")
        new = new.replace("NIVEL_ED_", "Educación: ")
        new = new.replace("ESTADO_", "Actividad: ")
        new = new.replace("CAT_INAC_", "Inactividad: ")
        if new == "CH06": new = "Edad (años)"
        if new == "IPCF": new = "Ingreso per cápita fam."
        new = limpiar_texto(new)
        rename_map[c] = new
    return X.rename(columns=rename_map)

def preparar_matriz(df_subset):
    dummies = pd.get_dummies(df_subset[cat_vars], drop_first=True, dtype=int)
    X = pd.concat([df_subset[num_vars], dummies], axis=1)
    X = renombrar_columnas(X)
    corr = X.corr(method="pearson")
    # TOP_K variables más relacionadas con (IPCF o Edad)
    objetivos = [c for c in ["Ingreso per cápita fam.", "Edad (años)"] if c in corr.columns]
    if objetivos:
        score = corr[objetivos].abs().max(axis=1).sort_values(ascending=False)
        keep = score.index[:min(TOP_K, len(score))]
        corr = corr.loc[keep, keep]
    return corr

def plot_heatmap_bonito(corr, titulo):
    mask = np.triu(np.ones_like(corr, dtype=bool))
    plt.figure(figsize=(14, 12))
    sns.heatmap(corr, mask=mask, cmap="coolwarm", vmin=-1, vmax=1, center=0,
                annot=False, square=True, cbar_kws={"shrink": 0.8},
                linewidths=0.3, linecolor="white")
    plt.title(titulo, fontsize=16, pad=12)
    plt.xticks(rotation=45, ha="right", fontsize=9)
    plt.yticks(rotation=0, fontsize=9)
    plt.tight_layout()
    plt.show()

# --- 6) Ejecutar por año (etiquetas ahora quedan consistentes) ---
for anio in ANIOS:
    df_a = df[df["ANO4"] == anio].copy()
    if df_a.empty:
        print(f"No hay datos para {anio}.")
        continue
    corr_a = preparar_matriz(df_a)
    plot_heatmap_bonito(corr_a, titulo=f"Matriz de correlación ({anio})")
```

```python
import pandas as pd
import numpy as np

# 1) Cargar base de hogares
df_hogar = pd.read_excel("df_hogar.xlsx")

# 2) Leer y preparar la tabla de adulto equivalente
tabla = pd.read_excel("tabla_adulto_equiv.xlsx", header=None)
tabla = tabla.iloc[4:28, :3]   # filas con edades y valores
tabla.columns = ["edad_label", "Mujer", "Varón"]

# Pasar a formato largo
tabla_long = tabla.melt(id_vars="edad_label", value_vars=["Mujer", "Varón"],
                        var_name="sexo", value_name="adulto_equiv")

# 3) Normalizar sexo en la base
def normalizar_sexo(x):
    s = str(x).strip().lower()
    if s in {"1","varon","varón"}: return "Varón"
    if s in {"2","mujer"}: return "Mujer"
    return np.nan

df_hogar["sexo_std"] = df_hogar["CH04"].apply(normalizar_sexo)

# 4) Pasar edad a tramo según tabla
def edad_a_tramo(e):
    if pd.isna(e): return np.nan
    try:
        e = int(np.floor(float(e)))
    except:
        return np.nan
    if e < 1: return "Menor de 1 año"
    if e == 1: return "1año"
    if 2 <= e <= 14: return f"{e} años"
    if e == 15: return "15 años"
    if e == 16: return "16 años"
    if e == 17: return "17 años"
    if 18 <= e <= 29: return "18 a 29 años"
    if 30 <= e <= 45: return "30 a 45 años"
    if 46 <= e <= 60: return "46 a 60 años"
    if 61 <= e <= 75: return "61 a 75 años"
    return "más de 75 años"

df_hogar["edad_label"] = pd.to_numeric(df_hogar["CH06"], errors="coerce").apply(edad_a_tramo)

# 5) Merge con la tabla para agregar adulto_equiv
df_hogar = df_hogar.merge(
    tabla_long,
    left_on=["sexo_std", "edad_label"],
    right_on=["sexo", "edad_label"],
    how="left"
).drop(columns=["sexo"])

# 6) Calcular suma por hogar (CODUSU + NRO_HOGAR)
df_hogar["ad_equiv_hogar3"] = df_hogar.groupby(["CODUSU","NRO_HOGAR"])["adulto_equiv"].transform("sum")

# >>> Ahora df_hogar tiene las columnas 'adulto_equiv' y 'ad_equiv_hogar3'
```

```python
# Valor de la canasta básica total por adulto equivalente
cba_2005 = 205.07
cba_2025 = 365177

# Crear columna ingreso_necesario según el año y el tamaño equivalente del hogar
df_hogar["ingreso_necesario"] = np.where(
    df_hogar["ANO4"] == 2005,
    df_hogar["ad_equiv_hogar3"] * cba_2005,
    df_hogar["ad_equiv_hogar3"] * cba_2025
)

df_hogar
```

```python
# 1) Columna pobre (1 si ITF < ingreso_necesario, 0 en caso contrario)
df_hogar["pobre"] = np.where(
    df_hogar["ITF"] < df_hogar["ingreso_necesario"], 1, 0
)

# 2) Cantidad de pobres por año
pobres_por_anio = df_hogar.groupby("ANO4")["pobre"].sum()

# 3) Total de observaciones por año
total_por_anio = df_hogar.groupby("ANO4")["pobre"].count()

# 4) Porcentaje de pobres
porcentaje_pobres = (pobres_por_anio / total_por_anio) * 100

# Mostrar resultados
print("Cantidad de pobres por año:")
print(pobres_por_anio)
print("\nPorcentaje de pobres sobre la muestra por año:")
print(porcentaje_pobres)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Crear columna 'pobre'
df_hogar["pobre"] = np.where(
    df_hogar["ITF"] < df_hogar["ingreso_necesario"], 1, 0
)

# 2) Estadísticas descriptivas por año
stats_pobre = df_hogar.groupby("ANO4")["pobre"].describe()
stats_pobre["suma"] = df_hogar.groupby("ANO4")["pobre"].sum()  # total de pobres
stats_pobre["porcentaje"] = df_hogar.groupby("ANO4")["pobre"].mean() * 100  # porcentaje de pobres

print("📊 Estadísticas descriptivas de 'pobre' por año")
print(stats_pobre)

# 3) Gráfico 1: porcentaje de pobres por año
porcentaje_pobres = df_hogar.groupby("ANO4")["pobre"].mean() * 100
plt.figure(figsize=(6,4))
porcentaje_pobres.plot(kind="bar", color=["skyblue", "salmon"])
plt.ylabel("% de pobres")
plt.title("Porcentaje de pobres por año")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Gráfico 2: distribución de pobres y no pobres por año
conteo = df_hogar.groupby(["ANO4","pobre"]).size().unstack(fill_value=0)

ax = conteo.plot(kind="bar", stacked=True, figsize=(6,4),
                 color=["lightgreen","tomato"])

plt.ylabel("Cantidad de personas")
plt.title("Distribución de pobres y no pobres por año")
plt.xticks(rotation=0)

# Ajustar la leyenda manualmente
plt.legend(["No pobre (0)", "Pobre (1)"], title="Condición")

plt.tight_layout()
plt.show()
```
